```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('Fifa 23 Players Data.csv')
df.info()
     
```

### 1. Membagi dataset menjadi training set dan testing set dengan proporsi 70:30

```python
X = df.iloc[:, : -1]
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

print("Dimensi X_train", X_train.shape)
print("Dimensi X_test", X_test.shape)
print("Dimensi y_train", y_train.shape)
print("Dimensi y_test", y_test.shape)
```

### 2. Melakukan normalisasi data pada salah satu attribute menggunakan Min Max scaler dengan mengcopy dataset terlebih dahulu

```python
min_max = MinMaxScaler()

# Copy Data
copy1 = df
dataCopy_normalization = copy1[["Overall"]]

normalization = min_max.fit_transform(dataCopy_normalization[["Overall"]])

df_normalization = pd.DataFrame(normalization)
df_normalization.columns =["Overall_Normalization"]

dataCopy_normalization = dataCopy_normalization.join(df_normalization["Overall_Normalization"]).reset_index()
dataCopy_normalization = dataCopy_normalization.drop("index", axis=1)

print()
print("Data Sebelum dan Sesudah dilakukan normalisasi : \n")
dataCopy_normalization.head(10)
```

### 3.Melakukan standarisasi pada dataset dengan membuat copy dataset terlebih dahulu

```python
data_copy = df.copy()

attribute_to_standardize = 'Value(in Euro)'

scaler = StandardScaler()

data_copy[attribute_to_standardize] = scaler.fit_transform(data_copy[[attribute_to_standardize]])

print("Atribut 'Value(in Euro) ' Sebelum dan Sesudah Standarisasi:\n")
print("Sebelum Standarisasi:\n")
print(df[[attribute_to_standardize]].head(10))
print("\nSesudah Standarisasi:\n")
print(data_copy[[attribute_to_standardize]].head(10))
```

### 4. Mengganti tipe data salah satu attribute angka

```python
df['Value(in Euro)'] = df['Value(in Euro)'].astype('float64')

print("Tipe data Value(in Euro) sesudah di ubah :")
df.dtypes
```

### 5. Lakukan Data cleaning pada data dengan nilai null dengan bilangan bulat : median/modus, bilangan desimal : mean, tulisan : modus

Lakukan pengecekan nilai null

```python
df.isna().sum()
```

Dikarenakan pada dataset tidak memiliki nilai null maka dibuat seperti diabawah

```python
pd.options.mode.chained_assignment = None

df.Potential[[729,3512,6892,5677,8931,9485,12047,14106,15271,17823]]=None;
df['Value(in Euro)'][[729,3512,6892,5677,8931,9485,12047,14106,15271,17823]]=None;
df['Best Position'][[729,3512,6892,5677,8931,9485,12047,14106,15271,17823]]=None;

df[df.isnull().any(axis=1)]
```

Data Bilangan Bulat dengan Strategi Median

```python
strategi1 = SimpleImputer(strategy="median")

df["Potential"] = strategi1.fit_transform(df[["Potential"]])

df.isna().sum()
```

```python
print(df.Potential[[729,3512,6892,5677,8931,9485,12047,14106,15271,17823]])
```

Data Bilangan Desimal dengan Strategi Mean

```python
strategi2 = SimpleImputer(strategy="mean")

df["Value(in Euro)"] = strategi2.fit_transform(df[["Value(in Euro)"]])

df.isna().sum()
```

```python
print(df['Value(in Euro)'][[729,3512,6892,5677,8931,9485,12047,14106,15271,17823]])
```

Data Tulisan dengan Strategi Modus

```python
strategi3 = SimpleImputer(missing_values=None, strategy="most_frequent")

df["Best Position"] = strategi3.fit_transform(df[["Best Position"]])

df.isna().sum()
```

```python
print(df['Best Position'][[729,3512,6892,5677,8931,9485,12047,14106,15271,17823]])
```
